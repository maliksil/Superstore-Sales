```python
df['Order Date']=pd.to_datetime(df['Order Date'],format='mixed')
df['Ship Date']=pd.to_datetime(df['Ship Date'],format='mixed')
```

```python
df.info()
```

```python
df.columns
```

```python
null_count=df['Postal Code'].isnull().sum()
print(null_count)
```

```python
# filling 0 to  the empty column
df['Postal Code'].fillna(0,inplace=True)
```

```python
df.info()
```

```python
#Changing from float to Integer.

df['Postal Code']=df['Postal Code'].astype(int)
df['Sales']=df['Sales'].astype(int)
```

```python
df.info()
```

```python
df.describe()
```

```python
# Data_Cleaning

# Checking Duplicate

if df.duplicated().sum() >0:
  print('duplicate are present')
else:
    print('no duplicate')
```

```python
df.duplicated().sum()
```

```python
# Exploratory Data Analysis.
# customer Segmentation

df.head(3)
```

```python
# type of cutomer

type_of_customer=df['Segment'].unique()
print(type_of_customer)
```

```python
# number of customer each segment

number_of_customer=df['Segment'].value_counts().reset_index()
number_of_customer=number_of_customer.rename(columns={'Segment':'Customer Type' })
print(number_of_customer)
```

```python
# Ploating A pie_Chart
plt.figure(figsize=(4,3))
plt.pie(number_of_customer['count'],labels=number_of_customer['Customer Type'],autopct='%1.1f%%',explode=[0,0,0.1])
plt.title('Distribution of Customer')
plt.show()
```

```python
# Customer And Sales

sales_per_catagory=df.groupby('Segment')['Sales'].sum().reset_index()
sales_per_catagory=sales_per_catagory.rename(columns={'Segment':'Customer Type' })
print(sales_per_catagory)
```

```python
plt.figure(figsize=(4,3))
plt.pie(sales_per_catagory['Sales'],labels=sales_per_catagory['Customer Type'],autopct='%1.1f%%',explode=[0,0,0.1])
plt.title('Sales per customer')
plt.show()
```

```python
# Bar_Graph

plt.figure(figsize=(4,3))
plt.bar(sales_per_catagory['Customer Type'],sales_per_catagory['Sales'])
plt.title('Sales per Catagory')
plt.xlabel('Customer type')
plt.ylabel('Sales')
plt.show()
```

```python
# Customer Loyality

customer_order_freq=df.groupby(['Customer ID','Customer Name','Segment'])['Order ID'].count().reset_index()
customer_order_freq=customer_order_freq.rename(columns={'Order ID':'Total Order'})
print(customer_order_freq)
```

```python
plt.figure(figsize=(4,3))
plt.bar(customer_order_freq['Segment'],customer_order_freq['Total Order'])
plt.title('Total Order per Segment')
plt.xlabel('Segment')
plt.ylabel('Total Order')
plt.show()
```
